```python
import pandas as pd
import os
import json
from pymongo import MongoClient
```

##2) Seteo de Path y archivos

```python
cwd = os.path.abspath(r'C:\Users\Lucia\Desktop\Facultad\UADE\202001_Diplomatura\00_twitter\DATA')
jsonFilePath = (cwd + '/rt_twits.json')
```

##3) Conexión con la base de datos y la colección

```python
client = MongoClient('mongodb://localhost:27017')
db = client['TPO_BigData']
cl = db['Twitter_RT']
```

##4) Carga de datos streaming en Mongo

```python
with open(jsonFilePath) as f:
    file_data = json.load(f)

if isinstance(file_data, list): 
    cl.insert_many(file_data)   
else: 
    cl.insert_one(file_data) 

client.close()
```
